```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/youtube/youtube.csv', index_col=0)
df
```

**Q1. Ïù∏Í∏∞ÎèôÏòÅÏÉÅ Ï†úÏûëÌöüÏàòÍ∞Ä ÎßéÏùÄ Ï±ÑÎÑê ÏÉÅÏúÑ 10Í∞úÎ™ÖÏùÑ Ï∂úÎ†•ÌïòÎùº(ÎÇ†ÏßúÍ∏∞Ï§Ä, Ï§ëÎ≥µÌè¨Ìï®)**

```python
Ans = list(df.loc[df['channelId'].isin(df['channelId'].value_counts().head(10).index)].channelTitle.unique())
Ans
```

**Q2. ÎÖºÎûÄÏúºÎ°ú Ïù∏Í∏∞ÎèôÏòÅÏÉÅÏù¥ Îêú ÏºÄÏù¥Ïä§Î•º ÌôïÏù∏ÌïòÍ≥† Ïã∂Îã§. dislikesÏàòÍ∞Ä likeÏàòÎ≥¥Îã§ ÎÜíÏùÄ ÎèôÏòÅÏÉÅ Ï±ÑÎÑêÏùÑ Î™®Îëê Ï∂úÎ†•ÌïòÎùº**

```python
Ans = list(df.loc[df['dislikes']>df['likes']].channelTitle.unique())
Ans
```

**Q3. Ï±ÑÎÑêÎ™ÖÏùÑ Î∞îÍæº ÏºÄÏù¥Ïä§Í∞Ä ÏûàÎäîÏßÄ ÌôïÏù∏ÌïòÍ≥† Ïã∂Îã§. channelIdÏùò Í≤ΩÏö∞ Í≥†Ïú†Í∞íÏù¥ÎØÄÎ°ú Ïù¥Î•º ÌÜµÌï¥ Ï±ÑÎÑêÎ™ÖÏùÑ ÌïúÎ≤àÏù¥ÎùºÎèÑ Î∞îÍæº Ï±ÑÎÑêÏùò Í∞ØÏàòÎ•º Íµ¨ÌïòÏó¨Îùº**

```python
change = df[['channelTitle', 'channelId']].drop_duplicates().channelId.value_counts()
Ans = change[change>1]
Ans
```

**Q4. ÏùºÏöîÏùºÏóê Ïù∏Í∏∞ÏûàÏóàÎçò ÏòÅÏÉÅÎì§ Ï§ë Í∞ÄÏû• ÎßéÏùÄ ÏòÅÏÉÅ Ï¢ÖÎ•ò(categoryId)Îäî Î¨¥ÏóáÏù∏Í∞Ä?**

```python
import datetime
df['trending_date2']=pd.to_datetime(df['trending_date2'])
Ans = df.loc[df['trending_date2'].dt.day_name()=='Sunday'].categoryId.value_counts().index[0]
#Ans = df.loc[df['trending_date2'].dt.weekday==6].categoryId.value_counts().index[0]
Ans
```

**Q5. Í∞Å ÏöîÏùºÎ≥Ñ Ïù∏Í∏∞ ÏòÅÏÉÅÎì§Ïùò categoryIdÎäî Í∞ÅÍ∞Å Î™áÍ∞úÏî©Ïù∏ÏßÄ ÌïòÎÇòÏùò Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑÏúºÎ°ú ÌëúÌòÑÌïòÎùº**

```python
group = df.groupby([df['trending_date2'].dt.day_name(), 'categoryId'], as_index=False).size()
Ans = group.pivot(index='categoryId', columns='trending_date2')
Ans
```

**Q6. ÎåìÍ∏ÄÏùò ÏàòÎ°ú (comment_count) ÏòÅÏÉÅ Î∞òÏùëÏóê ÎåÄÌïú ÌåêÎã®ÏùÑ Ìï† Ïàò ÏûàÎã§. viewcountÎåÄÎπÑ ÎåìÍ∏ÄÏàòÍ∞Ä Í∞ÄÏû• ÎÜíÏùÄ ÏòÅÏÉÅÏùÑ ÌôïÏù∏ÌïòÎùº. (view_countÍ∞íÏù¥ 0Ïù∏ Í≤ΩÏö∞Îäî Ï†úÏô∏)**

```python
target = df.loc[df['view_count'] != 0]
target['ratio'] = (target['comment_count']/target['view_count']).dropna()
Ans = target.sort_values(by='ratio', ascending=False).iloc[0].title
Ans
```

**Q6. ÎåìÍ∏ÄÏùò ÏàòÎ°ú(comment_count) ÏòÅÏÉÅ Î∞òÏùëÏóê ÎåÄÌïú ÌåêÎã®ÏùÑ Ìï† Ïàò ÏûàÎã§. viewcountÎåÄÎπÑ ÎåìÍ∏ÄÏàòÍ∞Ä Í∞ÄÏû• ÎÇÆÏùÄ ÏòÅÏÉÅÏùÑ ÌôïÏù∏ÌïòÎùº(view_countÍ∞íÏù¥ 0Ïù∏ Í≤ΩÏö∞Îäî Ï†úÏô∏)**

```python
target = (df['comment_count']/df['view_count']).dropna().sort_values()
target[target!=0].index[0]
Ans = df.loc[target[target!=0].index[0]].title
Ans
```

**Q8. like ÎåÄÎπÑ dislikeÏùò ÏàòÍ∞Ä Í∞ÄÏû• Ï†ÅÏùÄ ÏòÅÏÉÅÏùÄ Î¨¥ÏóáÏù∏Í∞Ä? (like, dislike Í∞íÏù¥ 0Ïù∏ Í≤ΩÏö∞Îäî Ï†úÏô∏)**

```python
target = df.loc[(df['likes']>0) & (df['dislikes']>0)]
num = (target['dislikes']/target['likes']).sort_values().index[0]
Ans = df.iloc[num].title
Ans
```
