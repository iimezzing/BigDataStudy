# 06. Reshape(Pivot)

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/pandas/main/under5MortalityRate.csv')
df
```

**Q83. Indicator을 삭제하고 First Tooltip 컬럼에서 신뢰구간에 해당하는 표현([~~])을 지우고 first 컬럼에 실수형으로 타입을 변경한 후 추가하라**

```python
df = df.drop(columns='Indicator')
df['First Tooltip'] = df['First Tooltip'].str.split('[').str[0].astype(float)
df
```

**Q84. 년도가 2015년 이상, Dim1이 Both sexes인 케이스만 추출**

```python
target = df[(df['Period']>=2015) & (df['Dim1']=='Both sexes')]
target
```

**Q85. 84번 문제에서 추출한 데이터로 아래와 같이 나라에 따른 년도별 사망률을 데이터 프레임화하기**

```python
Ans = target.pivot(index='Location', columns='Period', values='First Tooltip')
Ans
```

**Q86. Dim1에 따른 년도별 사망비율의 평균을 구하기**

```python
Ans = df.pivot_table(index='Dim1', columns='Period', values='First Tooltip', aggfunc='mean')
Ans
```

# 07. Merge , Concat

```python
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/pandas/main/mergeTEst.csv')
df
```

```python
df1 = df.iloc[:4,:]
df2 = df.iloc[4:,:]

display(df1)
display(df2)
```

**Q91. df1과 df2 데이터를 하나의 데이터 프레임으로 합치기**

```python
total = pd.concat([df1,df2])
total
```

```python
df3 = df.iloc[:2,:5]
df4 = df.iloc[5:,4:]

display(df3)
display(df4)
```

**Q92. df3과 df4 데이터를 하나의 데이터 프레임으로 합치기. 둘 다 포함하고있는 년도에 대해서만 고려**

```python
Ans = pd.concat([df3, df4], join='inner')
Ans
```

**Q93. df3과 df4 데이터를 하나의 데이터 프레임으로 합친다. 모든 컬럼을 포함하고 결측치는 0으로 대체**

```python
total1 = pd.concat([df3, df4]).fillna(0)
total1
```

```python
df5 = df.T.iloc[0:8,:3]
df6 = pd.concat([df.T.iloc[:1,2:5], df.T.iloc[7:,2:5]])
# 왜 기본적으로 인덱스가 생기는지 ㅠ 인덱스 없애는 방법을 찾아야할듯
df5.columns = df5.iloc[0]
df6.columns = df6.iloc[0]
display(df5)
display(df6)
```

**Q94. df5과 df6 데이터를 하나의 데이터프레임으로 merge함수를 이용하여 합치기. Algeria컬럼을 key로 하고 두 데이터 모두 포함하는 데이터만 출력**

```python
mer = pd.merge(df5, df6, on='Algeria', how='inner')
mer
```

**Q95. df5와 df6 데이터를 하나의 데이터 프레임으로 merge 함수를 이용하여 합치기. Algeria컬럼을 key로 하고 합집합으로 합치기**

```python
mer1 = pd.merge(df5, df6, on='Algeria', how='outer')
mer1
```

```python
# 숫자 형태로 변환
mer1['Afghanistan'] = pd.to_numeric(mer1['Afghanistan'])
# 'Afghanistan' 열을 숫자로 변환하고 변환할 수 없는 값은 NaN으로 처리
# mer1['Afghanistan'] = pd.to_numeric(mer1['Afghanistan'], errors='coerce')
# 'Afghanistan' 열을 기준으로 내림차순 정렬
mer1_sorted = mer1.sort_values(by='Afghanistan', ascending=False).reset_index(drop=True)

mer1_sorted
```
