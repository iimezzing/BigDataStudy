# 03_Grouping

**Q.44. 데이터를 로드하고 상위 5개 컬럼 출력**

```python
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/pandas/main/AB_NYC_2019.csv')
df.head()
```

**Q.45 데이터의 각 host_name의 빈도수를 구하고 host_name으로 정렬하여 상위 5개 출력**

```python
#Ans = df['host_name'].value_counts().sort_index().head()
Ans = df.groupby('host_name').size().head()
Ans
```

**Q46. 데이터의 각 host_name의 빈도수를 구하고 빈도수 기준 내림차순 정렬한 데이터 프레임 만들기. 빈도수 컬럼은 counts로 명명**

```python
#Ans = df.groupby('host_name').size().to_frame().rename(columns={0:'counts'}).sort_values('counts', ascending=False)
Ans = df['host_name'].value_counts().to_frame().rename(columns={0:'counts'})
Ans
```

**Q47. neighbourhood_group의 값에 따른 neighbourhood 컬럼 값의 개수 구하기**

```python
Ans = df.groupby(['neighbourhood_group','neighbourhood'],as_index=False).size()
Ans
```

**Q48. neighbourhood_group의 값에 따른 neighbourhood컬럼 값 중 neighbourhood_group그룹의 최댓값들을 출력**

```python
Ans = df.groupby(['neighbourhood_group','neighbourhood'],as_index=False).size().groupby(['neighbourhood_group'], as_index=False).max()
Ans
```

**Q49. neighbourhood_group 값에 따른 price 값의 평균, 분산, 최대, 최소 값 구하기**

```python
Ans = df[['neighbourhood_group','price']].groupby('neighbourhood_group').agg(['mean', 'var', 'max', 'min'])
# Ans = df[['neighbourhood_group','price']].groupby('neighbourhood_group').describe()
Ans
```

**Q50. neighbourhood_group 값에 따른 reviews_per_month 평균, 분산, 최대, 최소 값을 구하여라**

```python
Ans = df[['neighbourhood_group', 'reviews_per_month']].groupby('neighbourhood_group').agg(['mean', 'var', 'max', 'min'])
Ans
```
